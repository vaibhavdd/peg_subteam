Use != over <>

```python
with open('C:\\Users\\Animesh.Porwal//dim_customer.sql') as f:
    lines = f.readlines()
```

```python
lines
```

```python
lines_sign = []
for count, i in enumerate(lines):
    if i.__contains__("<>"):
        lines_sign.append([count,i])
```

```python
lines_sign
```

### Replace <> with != Rule


```python
for count, i in enumerate(lines_sign):
    print("!= not present in line {} we replaced".format(i[0]+1))
    lines_sign[count][1] = str(lines_sign[count][1]).replace('<>', '!=')
```
