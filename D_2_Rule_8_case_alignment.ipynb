Aligning case/when statements: each when should be on its own line (nothing on the case line) and should be indented one level deeper than the case line. The then can be on the same line or on its own line below it, just aim to be consistent.

```python
import numpy as np
import pandas as pd
import re
```

```python
with open('dim_customer.sql') as f:
    lines = f.readlines()
```

```python
def join_indentation(lines):
    lines_case = []
    lines_when = []
    lines_select = []
    select_when_pair = []
    select_case_pair = []
    
    for count, i in enumerate(lines):
        if i.__contains__("select"):
            lines_select.append([count,i])
        if i.__contains__("case"):
            lines_case.append([count,i])
        if i.__contains__(" when "):
            lines_when.append([count,i])
        if i.__contains__(" end as "):
            lines_when.append([count,i])
    
    for i in lines_when:
        j = i[0]
        while not lines[j].__contains__("select"):
            j-=1
        select_when_pair.append([i[0], j])

    for i in lines_case:
        j = i[0]
        while not lines[j].__contains__("select"):
            j-=1
        select_case_pair.append((j, i[0]))  
    
    for i in select_case_pair:
        case_indent = re.search(r'[a-z]', lines[i[1]], re.I).start()
        select_indent = re.search(r'[a-z]', lines[i[0]], re.I).start()
        if select_indent+4 != case_indent:
            print("Indentation of 'case' incorrect at line: {}".format(i[1]+1))
            
    for i in select_when_pair:
        when_indent = re.search(r'[a-z]', lines[i[0]], re.I).start()
        select_indent = re.search(r'[a-z]', lines[i[1]], re.I).start()
        if select_indent+8 != when_indent:
            print("Indentation of 'when' incorrect at line: {}".format(i[0]+1))


```

```python
join_indentation(lines)
```
